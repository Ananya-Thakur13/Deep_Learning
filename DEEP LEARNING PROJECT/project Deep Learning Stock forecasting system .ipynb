```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\PYTHON - TA\stock_train.csv")
df
```

```python
x = df["Open"]
x=np.array(x).reshape(len(x),1)  # 1 i column
print(x.shape)
```

```python
# min max scaling ________________________________________________________________
from sklearn.preprocessing import MinMaxScaler
scaler =MinMaxScaler()
x = scaler.fit_transform(x)
x
```

```python
x_stock=[]  # will go independent values here 
y_stock=[]   # dependent values here 

for i in range(30,len(x)):      # 10 is time stamp
    x_stock.append(x[i-30:i]) # same logic [1:11] -> [11] 1 se lekr 10 tk 11th ka data predict
    y_stock.append(x[i])
```

```python
print(x_stock,y_stock)
```

```python
import numpy as np
x_stock=np.array(x_stock)
y_stock=np.array(y_stock)
x_stock
```

```python
print(y_stock)
```

```python
print(type(x_stock))
x_stock
```

```python
# how to convert data into time series format ?
```

```python
# Training 
# data shape  ---> (rows,timestamp,feature)
# rows - no of samples 
# time stamp - no of days required to predict one stock
# Features - no of independent feature using to predict 
print(x_stock.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM,Dense,Dropout
from tensorflow.keras.models import Sequential
lstm = Sequential()
# first lstm layer
lstm.add(LSTM(units =50,input_shape=(x_stock.shape[1],x_stock.shape[2]),return_sequences=True))  # 1st layer of lstm, units = no of neurons
# return_sequences =True the lstm returns output at each time step, tis means the otp will be the sequence of outputs not just the final one
lstm.add(Dropout(0.30))
# second lstm layer 
lstm.add(LSTM(units=20))

# Output Layer
lstm.add(Dense(units=1,activation="relu"))  # relu bcs it is regression, we want numeric val not in form of 0 or 1
```

```python
# Compiling 
lstm.compile (optimizer="adam",loss="mean_squared_error")
```

```python
# fit
history=lstm.fit(x_stock,y_stock,batch_size=10,epochs=50)
import matplotlib.pyplot as plt
plt.plot(history.history["loss"])
plt.show()
```

```python
# prediction 
y_pred=lstm.predict(x_stock)
print(y_pred) # predicted value
```

```python
y_stock=scaler.inverse_transform(y_stock)
y_stock # actual value
```

```python
import matplotlib.pyplot as plt
plt.plot(y_stock, label='Actual Values')
plt.plot(y_pred, label='Predicted Values')

# Adding titles and labels
plt.title('Stock Price Prediction vs Actual Values')
plt.show
```

```python
last_30_days_pred = x[-30:]  # capturing last 30 days using slicing
print(last_30_days_pred)
```

```python
# we needa convert above var into 3d array for prediction in lstm
last_30_days_pred= np.reshape(last_30_days_pred,(1,30,1))
print(last_30_days_pred.shape)
```

```python
jan_1_2017= lstm.predict(last_30_days_pred)   # we have to predict for this date
jan_1_2017 = scaler.inverse_transform(jan_1_2017)
print(jan_1_2017)   # final predicted value
```

```python
# forecasting of stock price for next 10 days not just 1 day
predict = np.array([])
for i in range(10):    # to predict 10 days data
    d=x[-30:]
    next_day=lstm.predict(np.reshape(d,(1,30,1)))
    predict=np.append(predict,next_day)
    x=np.append(x,next_day)  # appending next_day data bcs to update time series format
```

```python
print(predict)
```

```python
predict=scaler.inverse_transform(np.reshape(predict,(10,1)))
print(predict)  # hence next 10 days prediction done 
```
